```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.datasets import load_iris
# load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

iris_df.head()
```

1. Compute mean, median, and mode for each feature.

```python
mean_values = iris_df.mean()
print(mean_values)
```

2. Calculate the range, variance, and standard deviation for each feature.

```python
range_values = iris_df.max() - iris_df.min()
print(range_values)
```

3. Compute the 25th, 50th, and 75th percentiles (quartiles) for each feature.

```python
quantile_values = iris_df.quantile([0.25, 0.5, 0.75])
print(quantile_values)
```
